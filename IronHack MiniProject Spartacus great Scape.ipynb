```python
from PIL import Image
import pygame
import cv2

class EscapeRoomGame:
    def __init__(self):
        # Initialize game state variables
        self.current_room = "Prison Cell"  # Start in the Prison Cell
        self.inventory = []  # Inventory to keep track of collected items
        
        # Define room descriptions, items, and interactions
        self.rooms = {
            "Prison Cell": {
                "description": "You wake up on the cold stone floor of your cell, bound and bruised. The only light comes from a small, barred window high above. As your eyes adjust to the darkness, you notice a loose stone in the wall.",
                "items": {"loose stone": "A loose stone in the wall."},
                "interactions": {"loose stone": "You push the stone aside and find a small, rusted key."}
            },
            "Library": {
                "description": "You step into the first room, a large library filled with ancient scrolls and dusty tomes. The room is dimly lit, and the scent of old parchment fills the air. To your right, there is a heavy door, and to your left, a desk cluttered with scrolls and a locked drawer.",
                "items": {"desk": "A desk cluttered with scrolls and a locked drawer.", "scrolls": "Ancient scrolls are scattered across the desk."},
                "interactions": {"desk": "You find a small vial of oil among the scrolls."}
            },
            "Garden": {
                "description": "You enter the next room, a serene garden filled with the sounds of trickling water and birdsong. The floor is a mosaic of colored tiles, and every step must be taken with caution. In the center of the garden, there is a fountain with a locked box, and near the entrance, a vine-covered archway leading to a hidden alcove.",
                "items": {"tiles": "A mosaic of colored tiles.", "fountain": "A fountain with a locked box.", "archway": "A vine-covered archway."},
                "interactions": {"tiles": "You notice a pattern that resembles constellations. Following this path avoids the traps."}
            },
            "Workshop": {
                "description": "After narrowly escaping the garden, you find yourself in a workshop filled with tools and gadgets. The walls are lined with shelves holding various mechanical parts.",
                "items": {"shelves": "Shelves filled with tools, gears, and assorted mechanical parts.", "tools": "Various tools are scattered on the workbenches."},
                "interactions": {"shelves": "You find a dusty, old blueprint detailing a complex locking mechanism."}
            },
            "Study": {
                "description": "You step into the last room, a grand study with high walls filled with books and scrolls. At the far end, you see a table with various models of screens and an open book describing their specifications.",
                "items": {"bookshelves": "Bookshelves filled with ancient tomes, manuscripts, and scrolls.", "table": "A table with various models of screens and an open book."},
                "interactions": {"bookshelves": "You find a book titled 'Ancient Mechanisms' that seems relevant to the workshop diagram."}
            }
        }
        
        # Define image paths for rooms
        self.images = {
            "Library": "library.jpg",
            "Garden": "garden.jpg",
            "Workshop": "workshop.jpg",
            "Study": "study.jpg"
        }
        
        # Define video paths for rooms
        self.videos = {
            "Prison Cell": r"C:\Users\Admin\OneDrive\Documentos\IronHack\Mini_Project_Game_Spartacus_great_Escape\Video\Awakening.mp4"
        }
        
        # Initialize pygame for sound playback
        pygame.init()
        self.sounds = {
            "prison_cell": r"C:\Users\Admin\OneDrive\Documentos\IronHack\Mini_Project_Game_Spartacus_great_Escape\Sounds\prison_cell.mp3",
            "library": r"C:\Users\Admin\OneDrive\Documentos\IronHack\Mini_Project_Game_Spartacus_great_Escape\Sounds\library.mp3",
            "garden": r"C:\Users\Admin\OneDrive\Documentos\IronHack\Mini_Project_Game_Spartacus_great_Escape\Sounds\garden.mp3",
            "workshop": r"C:\Users\Admin\OneDrive\Documentos\IronHack\Mini_Project_Game_Spartacus_great_Escape\Sounds\workshop.mp3",
            #"study": "study.mp3",
            #"unlock": "unlock.mp3",
            #"locked": "locked.mp3"
        }

    def start(self):
        # Print welcome message and start the game by showing the first room
        print("Welcome to the Escape Room!")
        self.show_room()

    def show_room(self):
        # Display the current room's description and media (image or video)
        room = self.current_room
        print("\n" + self.rooms[room]["description"])

        # Play the sound for the current room
        #self.play_sound(self.sounds[room.lower()]) 

        # Show image if the room has an associated image

        if room in self.videos:
            self.play_video(self.videos[room])
        
       
        # Show options for the player
        self.show_options()

    def show_options(self):
        # Display options for the player to choose from
        print("\nOptions:")
        print("1. Explore the room")
        print("2. Examine an item")
        print("3. Interact with an item")
        choice = input("Enter the number of your choice: ")
        
        # Call the appropriate method based on the player's choice
        if choice == "1":
            self.explore()
        elif choice == "2":
            self.examine()
        elif choice == "3":
            self.interact()
        else:
            print("Invalid choice.")
            self.show_options()

    def explore(self):
        # Provide an overview of the current room
        room = self.current_room
        print("\nExploring the room:")
        print(self.rooms[room]["description"])
        self.show_options()

    def examine(self):
        # Allow the player to examine items in the current room
        room = self.current_room
        print("\nItems to examine:")
        
        # List all items available for examination in the room
        for item in self.rooms[room]["items"]:
            print(f"- {item}")
        
        # Get the player's choice and provide the item's description
        choice = input("Enter the item you want to examine: ").strip().lower()
        if choice in self.rooms[room]["items"]:
            print(self.rooms[room]["items"][choice])
        else:
            print("Invalid item.")
        
        # Show options again
        self.show_options()

    def interact(self):
        # Allow the player to interact with items in the current room
        room = self.current_room
        print("\nItems to interact with:")
        
        # List all items available for interaction in the room
        for item in self.rooms[room]["interactions"]:
            print(f"- {item}")
        
        # Get the player's choice and perform the interaction
        choice = input("Enter the item you want to interact with: ").strip().lower()
        if choice in self.rooms[room]["interactions"]:
            print(self.rooms[room]["interactions"][choice])
            self.handle_interaction(choice)
        else:
            print("Invalid item.")
        
        # Show options again
        self.show_options()

    def handle_interaction(self, choice):
        # Handle the interaction logic and update the game state accordingly
        if self.current_room == "Prison Cell" and choice == "loose stone":
            self.inventory.append("rusted key")  # Add key to inventory
            self.current_room = "Library"  # Move to the next room
        elif self.current_room == "Library" and choice == "desk":
            self.inventory.append("vial of oil")  # Add vial of oil to inventory
        elif self.current_room == "Garden" and choice == "tiles":
            self.current_room = "Workshop"  # Move to the next room
        elif self.current_room == "Workshop" and choice == "shelves":
            self.inventory.append("blueprint")  # Add blueprint to inventory
        elif self.current_room == "Study" and choice == "bookshelves":
            self.inventory.append("ancient mechanisms book")  # Add book to inventory
            self.current_room = "Garden"  # Move to the next room
        
        # Show the updated room
        self.show_room()

    def show_image(self, image_path):
        # Display the image using PIL
        try:
            img = Image.open(image_path)
            img.show()
        except Exception as e:
            print(f"Error: {e}")

    def play_video(self, video_path):
        # Play the video using OpenCV
        cap = cv2.VideoCapture(video_path)
        if not cap.isOpened():
            print(f"Error: Could not open video {video_path}")
            return

        while cap.isOpened():
            ret, frame = cap.read()
            if not ret:
                break
            cv2.imshow('Prison Cell', frame)
            if cv2.waitKey(25) & 0xFF == ord('q'):
                break

        cap.release()
        cv2.destroyAllWindows()

    def play_sound(self, sound_path):
        # Play the sound using pygame
        try:
            pygame.mixer.music.load(sound_path)
            pygame.mixer.music.play()
        except Exception as e:
            print(f"Error: {e}")

if __name__ == "__main__":
    # Start the game
    game = EscapeRoomGame()
    game.start()
```
